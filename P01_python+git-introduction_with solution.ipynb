```python
import numpy as np
import math 

x = np.linspace(0.0,1.0,10000) 

theta=[]
phi=[]

for idx,i in enumerate(x,1):
    
    Theta = 2 * math.pi  * i   
    Phi = math.acos( 2 * i - 1 ) 
    theta.append(Theta)
    phi.append(Phi)
    
    #print( "{}'th point has angular coordinates : theta = {}, phi = {}".format(idx,Theta,Phi))    # To see the coordinates we need to activiate this code
```

```python
# Assumption : unit radius
x = np.sin(phi) * np.cos(theta)
y = np.sin(phi) * np.sin(theta)
z = np.cos(phi)

#for a,b,c in zip(x,y,z) :
    #print( "Carthesian coordinates : x = {}, y = {}, z = {}".format(a,b,c))  # To see the coordinates we need to activiate this code
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))

# Plot the projection on the x-axis
plt.subplot(131)
plt.scatter(y, z)
plt.title('Projection on the yz-plane')
plt.xlabel('y')
plt.ylabel('z')
plt.axis('equal')

# Plot the projection on the y-axis
plt.subplot(132)
plt.scatter(x, z)
plt.title('Projection on the xz-plane')
plt.xlabel('x')
plt.ylabel('z')
plt.axis('equal')

# Plot the projection on the z-axis
plt.subplot(133)
plt.scatter(x, y)
plt.title('Projection on the xy-plane')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')

plt.tight_layout()
```

```python
"""
bins : This parameter specifies the number of intervals into which the data should be divided for the histogram. 

alpha : This parameter controls the transparency of the histogram bars. 
A value of 1 would make them fully opaque, while a value of 0 would make them fully transparent.

"""


plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.hist(x, bins=50, alpha=0.70, color='blue')
plt.title('X Distribution')
plt.xlabel('x')
plt.ylabel('Frequency')

plt.subplot(132)
plt.hist(y,bins=50, alpha=0.70, color='green')
plt.title('Y Distribution')
plt.xlabel('y')
plt.ylabel('Frequency')

plt.subplot(133)
plt.hist(z, bins=50, alpha=0.70, color='red')
plt.title('Z Distribution')
plt.xlabel('z')
plt.ylabel('Frequency')

plt.tight_layout()
```
