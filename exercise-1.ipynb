```python
import numpy as np
```

```python
#first task: Write a piece of code that will sample 10,000 points uniformly distributed on the surface of a sphere. Your code should return the points' angular coordinates  ùúÉ,ùúô (you can assume a unit sphere wlog).

def sample_sphere(n):

    phi=2*np.pi*np.random.rand(n)
    theta=np.pi*np.random.rand(n) #if we define theta this way, the sample will not be uniformly ditribute. 
    
    return theta, phi

theta, phi = sample_sphere(10000)
print(theta,phi)
```

```python
#second task: Write a piece of code that will project the points into Carthesian coordinates (ùë•,ùë¶,ùëß), again assuming unit radius.

#with radius r=1

def proj_on_cart(theta,phi):
    x=np.sin(theta)*np.cos(phi)
    y=np.sin(theta)*np.sin(phi)
    z=np.cos(theta)
    
    return x, y, z

x,y,z = proj_on_cart(theta,phi)
print(x,y,z)
```

```python
#third task: Make three 2D scatter plots of the points after projecting along the ùë•,ùë¶ and ùëß axis.

import matplotlib.pyplot as plt

theta, phi = sample_sphere(10000)
x, y, z = proj_on_cart(theta, phi)

# Third task: Scatter plots of projections
plt.figure(figsize=(15, 5))

# Project along x-axis
plt.subplot(1, 3, 1)
plt.scatter(y, z, s=1)
plt.title("Projection along X-axis")
plt.xlabel("y")
plt.ylabel("z")

# Project along y-axis
plt.subplot(1, 3, 2)
plt.scatter(x, z, s=1)
plt.title("Projection along Y-axis")
plt.xlabel("x")
plt.ylabel("z")

# Project along z-axis
plt.subplot(1, 3, 3)
plt.scatter(x, y, s=1)
plt.title("Projection along Z-axis")
plt.xlabel("x")
plt.ylabel("y")

plt.tight_layout()
plt.show()
```

```python
#4th task: Plot a 1D histogram of the ùë•,ùë¶ and ùëß distributions.

# First task: Sample points
theta, phi = sample_sphere(10000)

# Second task: Convert to Cartesian coordinates
x, y, z = proj_on_cart(theta, phi)

# Fourth task: 1D histograms for x, y, and z distributions
plt.figure(figsize=(15, 5))

# Histogram for x distribution
plt.subplot(1, 3, 1)
plt.hist(x, bins=50, edgecolor='black', alpha=0.7)
plt.title("X Distribution")
plt.xlabel("x")
plt.ylabel("Frequency")

# Histogram for y distribution
plt.subplot(1, 3, 2)
plt.hist(y, bins=50, edgecolor='black', alpha=0.7)
plt.title("Y Distribution")
plt.xlabel("y")
plt.ylabel("Frequency")

# Histogram for z distribution
plt.subplot(1, 3, 3)
plt.hist(z, bins=50, edgecolor='black', alpha=0.7)
plt.title("Z Distribution")
plt.xlabel("z")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
